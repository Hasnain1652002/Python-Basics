```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = '''Apple Inc. was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne in 1976. 
The company, headquartered in Cupertino, California, revolutionized personal computing with the Macintosh in 1984. 
As of 2024, Apple is one of the most valuable companies in the world, with a market capitalization of over $2 trillion.'''

doc = nlp(text)

sentences = [sent.text for sent in doc.sents]
words = [token.text for token in doc]

words_without_stopwords = [token.text for token in doc if not token.is_stop]

lemmatized_words = [token.lemma_ for token in doc]

named_entities = [(ent.text, ent.label_) for ent in doc.ents]

pos_tags = [(token.text, token.pos_) for token in doc]
```

```python
print("\nTokenized Sentences:")
print("-" * 25)
for i, sent in enumerate(sentences, 1):
    print(f"{i}. {sent}")
```

```python
print("\nTokenized Words:")
print("-" * 25)
print(", ".join(words))
```

```python
print("\nWords without Stopwords:")
print("-" * 25)
print(", ".join(words_without_stopwords))
```

```python
print("\nLemmatized Words:")
print("-" * 25)
print(", ".join(lemmatized_words))
```

```python
print("\nNamed Entities and Labels:")
print("-" * 25)
for ent, label in named_entities:
    print(f"{ent} -> {label}")
```

```python
print("\nPart of Speech Tags:")
print("-" * 25)
for word, pos in pos_tags:
    print(f"{word}: {pos}")
```
